```python
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('./genetics.csv')
Xb = sm.add_constant(df[['is_oncology', 'has_genetics', 'is_orphan']])
mod = sm.OLS(df.expedited, Xb)
res = mod.fit()
res.summary()
```
