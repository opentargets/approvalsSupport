```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3
import seaborn as sns
```

```python
Expedited = 287
Severe = 289
GE = 271

Expedited_Severe = 233
Expedited_GE = 181
Severe_GE = 201

Expedited_Severe_GE = 161

a = Expedited
b = Severe
c = GE

ab = Expedited_Severe
bc = Severe_GE
ca = Expedited_GE

abc = Expedited_Severe_GE

A = a - ab - ca + abc
B = b - ab - bc + abc
C = c - ca - bc + abc

AB = ab - abc
BC = bc - abc
CA = ca - abc

ABC = abc
```

```python
# Expedited_Severe/Expedited * 100
Severe_GE/Severe * 100
```

```python
palette = sns.color_palette('Set2') 
```

```python
plt.rcParams['font.family'] = 'Helvetica'
```

```python
plt.figure(dpi=300)
prod = (A, B, AB, C, CA, BC, ABC)
venn3(subsets = prod, 
      set_labels = ('Expedited', 'Serious', 'Genetic evidence'),
      set_colors=palette[:3])
plt.show()
```
